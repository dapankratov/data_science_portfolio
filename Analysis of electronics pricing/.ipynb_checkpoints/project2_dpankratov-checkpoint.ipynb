Data set has 2 columns max and min price. I will create mean price column which is going to be used in the analysis.

```python
df['prices.amountMean'] = df[['prices.amountMax', 'prices.amountMin']].mean(axis=1)
```

```python
df.dtypes
```

The column prices.dateSeen contains data on different years and time a retailer has the item on sale with a specific price. To make a more correct analysis we need to compare prices on the same year. I will use only 2018. 
Additionally we can see that sometimes the column has more than one time period. I will use only first occurrence.

```python
df['prices.dateSeen'].value_counts()
```

```python
#spliting strings by ',' char and overwriting the column with first element giving it
#datetime type than recieving only year using 'dt' call
df['prices.dateSeen'] = df['prices.dateSeen'].str.split(pat = ",").str[0].astype('datetime64[ns]').dt.year 
```

## Hypothesis 2
<p>h0: Retail price is not correlated with a number of stores retailer has in USA.<br>
h1: There is a positive correlation between prices and the number of stores that a retailer has in the USA.</p>
<p>To check the hypothesis I will analyze the mean prices for the top 2 retailers (Bestbuy.com, bhphotovideo.com) because they have a similar amount of data accessible to me.</p>

```python
from bs4 import BeautifulSoup
import requests as req
```

```python
resp = req.get("https://stores.bestbuy.com/")
soup = BeautifulSoup(resp.text, 'html')
```

```python
x = soup.find(id='js-directory-locationCount').text
```

```python
print('Bestbuy.com has ' + x + ' locations.')
```

```python
resp = req.get("https://en.wikipedia.org/wiki/B%26H_Photo")
soup = BeautifulSoup(resp.text, 'html')
```

```python
#Td which I need can't be accessed directly because there are many Tds on the page. This is why I used unique Th to help with it 
x = soup.find('th', text="Number of locations").find_next_sibling("td").text
```

```python
print('Bhphotovideo.com has ' + x + ' location.')
```

<b>We can clearly see that number of stores a retailer has does not affect the prices of items (the describution of prices was plotted earlier). It means that H0 for Hypothesis 2 is correct. </b>
