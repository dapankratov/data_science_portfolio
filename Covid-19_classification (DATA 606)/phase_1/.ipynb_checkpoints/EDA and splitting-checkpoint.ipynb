### Preparation of Covid-positive working dataset

```python
'''
This code finds all images of patients of a specified VIRUS and X-Ray view and stores selected image to an OUTPUT directory
It uses metadata.csv for searching and retrieving images name
Using ./images folder it selects the retrieved images and copies them in output folder
Code can be modified for any combination of selection of images
'''

# Selecting all combination of 'COVID-19' patients with 'PA', 'AP' and 'AP Supine' X-Ray view
virus = "Pneumonia/Viral/COVID-19" # Virus to look for

metadata = "covid-chestxray-dataset/metadata.csv" # Meta info
imageDir = "covid-chestxray-dataset/images" # Directory of images
outputDir = "data/covid_positive/" # Output directory to store selected images

try:
    if not os.path.isfile(outputDir):  # check if directory already exist
        os.mkdir(outputDir)  # create a directory
except OSError:
    pass

metadata_csv = pd.read_csv(metadata)

# loop over the rows of the COVID-19 data frame
for (i, row) in metadata_csv.iterrows():
	if row["finding"] != virus or ( row["view"] != 'PA' and row["view"] != 'AP' and row["view"] != 'AP Supine'):
		continue

	filename = row["filename"].split(os.path.sep)[-1]
	filePath = os.path.sep.join([imageDir, filename])
	copy2(filePath, outputDir)
```

```python
# Plotting of a sample of COVID-positive dataset
fig, axes = plt.subplots(figsize=(12,10), nrows=2, ncols=2)
fig.suptitle('Covid positive', fontsize=16)

image_paths = glob('data/covid_positive/*')
for imp, ax in zip(image_paths, axes.ravel()):
    img = cv2.imread(imp)
    ax.imshow(img)

#plt.savefig('images/covid_positive.png')
```

## Splitting prepared images into Train/Val/Test folders 

```python
#Thank you https://medium.com/@Ana_Caballero_H/splitting-data-folders-into-training-validation-and-testing-folders-using-shuffling-with-python-63cabcd997dd

data_dir='data'
print(len(os.listdir('data/covid_positive')))
print(len(os.listdir('data/covid_negative')))
```

```python
classes = os.listdir(data_dir)
print(classes)
```

```python
source_path = [f'data/{a}' for a in classes]
print(source_path)
```

```python
try:
    os.mkdir('data/train')
    os.mkdir('data/val')
    os.mkdir('data/test')
except OSError:
    pass
```

```python
TRAINING_PATH='data/train'
VALIDATION_PATH='data/val'

training_dir_path = [f'data/train/{a}' for a in classes]
print('\n',training_dir_path)

validation_dir_path = [f'data/val/{a}' for a in classes]
print('\n',validation_dir_path)

testing_dir_path = [f'data/test/{a}' for a in classes]
print('\n',testing_dir_path)
```

```python
for train_dir_path in training_dir_path:
    try:
        os.mkdir(train_dir_path)
    except OSError:
        pass
```

```python
for val_dir_path in validation_dir_path:
    try:
        os.mkdir(val_dir_path)
    except OSError:
        pass
```

```python
for test_dir_path in testing_dir_path:
    try:
        os.mkdir(test_dir_path)
    except OSError:
        pass
```

```python
def split_data(SOURCE, TRAINING, VALIDATION, TESTING, SPLIT_SIZE):
    files = []
    print('Split Data')
    for filename in os.listdir(SOURCE):
        file = SOURCE +'/'+ filename
        if os.path.getsize(file) > 0:
            files.append(filename)
        else:
            print(filename + " is zero length, so ignoring.")

    training_length = int( len(files)* SPLIT_SIZE)
    validation_length = int(len(files) * 0.10)
    testing_length = int(len(files) - training_length - validation_length)
    
    print('SOURCE: ',SOURCE, '\n TRAINING', TRAINING, '\n VALIDATION',VALIDATION, '\n ',len(files))
    print('training_length:',training_length)
    print('validation_length:',validation_length)
    print('testing_length:',testing_length)
    
    shuffled_set = random.sample(files, len(files))
    training_set = shuffled_set[0:training_length]
    validation_set = shuffled_set[training_length:(training_length+validation_length)]
    testing_set = shuffled_set[:testing_length]

    print(len(training_set))
    print(len(validation_set))
    print(len(testing_set))
    
    for filename in training_set:
        this_file = SOURCE +'/'+ filename
        destination = TRAINING +'/'+ filename
        copy2(this_file, destination)
    
    for filename in validation_set:
        this_file = SOURCE +'/'+ filename
        destination = VALIDATION +'/' + filename
        copy2(this_file, destination)
        
    for filename in testing_set:
        this_file = SOURCE +'/'+ filename
        destination = TESTING +'/' + filename
        copy2(this_file, destination)
```

```python
split_size = .85
for source,train_dir_path,val_dir_path,test_dir_path in zip(source_path,\
                                training_dir_path,validation_dir_path, testing_dir_path):
    #print('source: ',source,'\n', train_dir_path,'\n',val_dir_path,'\n')
    split_data(source,train_dir_path,val_dir_path,test_dir_path, split_size)
    print('Splitting \n')
```
